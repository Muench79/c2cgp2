# Tutorial Property

Beispielklasse Employee mit ungeregeltem Zugriff auf das Attribut `name`

```python
class Employee1:
    def __init__(self, name):
        self.name = name 

    def __repr__(self):
        return f"EMPLOYEE | NAME: {self.name}"
```

```python
x1 = Employee1('Max')
print(x1)
x1.name = 1000
print(x1)
```

```python
dir(x1)
```

Verwendung von Methoden zur Regelung des Zugriffs auf `name`

```python
class Employee2:
    def __init__(self, name):
        self.set_name(name)        # Verwendung einer Setter-Funktion

    def __repr__(self):
        return f"EMPLOYEE | NAME: {self._name}"

    def set_name(self,name):
        """setter methode"""
        if isinstance(name,str):
            self._name = name.upper() # Unterstricht deutet an, dass es ein "Protected Attribut" ist 
        else:
            raise TypeError

    def get_name(self):
        return self._name
```

```python
x2 = Employee2('Max')
print(x2)
x2.set_name('klaus')
print(x2)

try:
    x2.set_name(12)
except TypeError:
    print('! Achtung TypeError !')
print(x2)
```

```python
x2.get_name()
```

```python
dir(x2)
```

## Verwendung als Property
- Erlaubt die Verwendung des Zuweisungsoperator `=` mit geregeltem Zugriff


### Unter Verwendung der BuiltIn-funktion `property`

```python
class Employee3:
    def __init__(self, name):
        self.name = name  

    def __repr__(self):
        return f"EMPLOYEE | NAME: {self._name}"

    def set_name(self,name):
        """setter methode"""
        if isinstance(name,str):
            self._name = name.upper()
        else:
            raise TypeError

    def get_name(self):
        return self._name

    name = property(get_name,set_name)
```

```python
x3 = Employee3('Max')
print(x3)
x3.name='klaus'
print(x3)
try:
    x3.name=12
except TypeError:
    print('! Achtung TypeError !')
print(x3)
```

### Verwendung mit `@`

```python
class Employee4:
    def __init__(self, name):
        self._name = name  # Unterstricht deutet an, dass es ein "Protected Attribut" ist 

    def __repr__(self):
        return f"EMPLOYEE | NAME: {self._name}"

    @property
    def name(self):
        "getter"
        return self._name

    @name.setter
    def name(self,name):
        """setter"""
        if isinstance(name,str):
            self._name = name.upper()
        else:
            raise TypeError
```

```python
x4 = Employee4('Max')
print(x4)
x4.name='klaus'
print(x4)
try:
    x4.name=12
except TypeError:
    print('! Achtung TypeError !')
print(x4)
```
