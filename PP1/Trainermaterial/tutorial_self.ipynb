## Instance methodes

```python
class Myclass:
    
    def __init__(self,name):
        self._name = None
        self.set_name(name)
        #self.get_name = 'sdfsdf'

    def set_name(self,name:str):
        """sets attribute name and converts all letter to capital letters"""
        if isinstance(name,str):
        	self._name = name.upper()
        else:
            raise TypeError()
        
    def get_name(self)-> str:
        "Returns name"
        return self._name
    
    def info(self):
        return self
```

```python
x = Myclass("as")
```

```python
# normal call to method
x.get_name()
```

## Instance methodes, Class methodes and static methodes

```python
class Myclass2:
    
    # Instance method
    # First is reference to instance
    def func1(self): # Myclass2.func(instance of Myclass2)
        print(self)
    
    # Class method
    # First argument is reference to class   
    @classmethod
    def func2(cls): # Myclass2.func2(-Myclass2-)
        print(cls)
    
    # Static method
    # not special first argument
    @staticmethod
    def func3(p):   # Myclass2.func3()
        print(p)
        
# Myclass2.func2 = classmethod(Myclass.func2)
# Myclass2.func3 = staticmethod(Myclass.func3)
```

```python
x = Myclass2()
```

## Use cases

**class methodes**

- provide simplified access to the class itself with reference to an instance
- alternativ constructor method to create an instance from some information (can be inherited)

**static methode**
- no access to instance specific data
- utility functions related to a class
- often difficult to handle

```python
class Date(object):
    
    def __init__(self, day=0, month=0, year=0):
        self.day = day
        self.month = month
        self.year = year
        
    def __str__(self): # Magic method or Dunder-method interacts with built-in function str
        """creates string from instance"""
        return f"{self.day}-{self.month}-{self.year}"
    
    def __repr__(self):
        return 'ASDASD'

    @classmethod
    def from_string(cls, date_as_string):
        """creates new instance from string"""
        day, month, year = map(int, date_as_string.split('-'))
        date1 = cls(day, month, year)
        return date1

    @staticmethod
    def is_date_valid(date_as_string):
        """checks if string is a valid date string"""
        day, month, year = map(int, date_as_string.split('-'))
        return day <= 31 and month <= 12 and year <= 3999

if Date.is_date_valid('11-09-2012'):
    date = Date.from_string('11-09-2012')
print(date)
```
