# Tutorial Access Modifiers

#### Generelle Idee der Access Modifiers

**Public Access**: Auf das Attribute kann von jeder Stelle eines Programms zugegriffen werden.

**Protected**: Auf das Attribute kann nur aus der Klasse oder einer abgeleiteten Klasse zugegriffen werden.

**Private Access**: Auf das Attribute kann nur aus der Klasse zugegriffen werden. -> Das Attribute wird (unter diesem Namen) nicht vererbt.

Dieses Konzept wird in unterschiedlichen Programmiersprachen unterschiedlich umgesetzt!

```python
class A():
    def __init__(self,x,y,z):
        self.x = x      # public attribute
        self._y = y     # protected attribute (name mangling)
        self.__z = z    # private attribute ()

a = A(1,2,3)
```

```python
dir(a) # all attribute are available, but name of private is changed/mangled
```

```python
# access to public attribute x
print(a.x)
a.x = 10
print(a.x)
```

```python
# access to private attribute
print(a._y)
a._y = 20
print(a._y)
```

```python
# access to protected is not possible, since its name is mangled
print(a.__z)
```

```python
# access to mangled name is possible
a._A__z # renamed protected attribute
```

```python
# creation of new (other) attribute is possible
a.__z = 30
print(a.__z)
```

## Inheritance of protected attributes

```python
class B(A):
    pass # 

b = B(1,2,3)

b.x, b._y, b._A__z
```

```python
dir(b) # mangled name of protected attribute refered to class A not B
```

```python
class C(A):
    def __init__(self,x,y,z):
        #self.__z = z
        super().__init__(x,y,z)
        pass
        

c = C(1,2,3)

c.x, c._y, c._A__z, c._C__z
```

```python
dir(c)
```

## class attribute and methods follow the same rules

```python
class D:
    a = "public class attribute"
    _b = "protected class attribute"
    __c = "private class attribute"

    def fa(self):
        pass

    def _fb(self):
        pass

    def __fc(self):
        pass
```

```python
d = D()
dir(d)
```

```python
class E(D):
    pass
```
