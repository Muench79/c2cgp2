## Typisierung der Argumente und des Rückgabewertes von Funktionen

-	Statische vs. dynamische Typisierung! 
-	Python ist eine Sprache mit dynamischer Typisierung- -> Typisierungfehler in werden erst zur Laufzeit erkannt.
- 	Annotationen erlauben das Beschreiben/Festlegen der Typen
-	Diverse Tools erlauben das Prüfen auf Typisierungsfehler
-	Programme können so gestaltet werden, dass sie den Rückgabewert einer Funktion anpassen

```python
# Typen der Argumente kommentieren
def func(x: str, times : int = 10):
    return x*times

func('A')
```

```python
func.__annotations__
```

```python
# Typen der Argumente und des Rückgabewertes kommentieren (nur einer)
def func(x: str, times : int = 10) -> str:
    return x*times

func('A')
```

```python
func.__annotations__
```

```python
class A:
    def __init__(self,x:int):
        self._x = x
    def add(self,x:int) -> int:
        return self._x+x
```

```python
A.add.__annotations__
```

Annotationen sind KEINE Typenprüfung!

```python
A('asd')
```
