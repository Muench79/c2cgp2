```python
import pandas as pd
from dash import Dash, html, dcc, Input, Output, dash_table
import plotly.express as px
import numpy as np
```

```python
Dashdf = pd.read_csv("fahrdaten.csv")

#Optional timestampspalte in Zeitwerte konvertieren
Dashdf["timestamp"] = pd.to_datetime(Dashdf["timestamp"], unit="s")
Dashdf["fahrzeit_s"] = (Dashdf["timestamp"] - Dashdf["timestamp"].iloc[0]).dt.total_seconds()
Dashdf
```

```python
max_speed=Dashdf["speed"].max()
min_speed=Dashdf["speed"].min()
avg_speed=Dashdf["speed"].mean()
total_dist=(Dashdf["speed"] * Dashdf["fahrzeit_s"]).sum()
s = Dashdf["timestamp"].dropna().sort_values()                      # Werte sortieren
total_time=s.iloc[-1] - s.iloc[0]
total_time
```

```python
app = Dash(__name__)

# App layout
app.layout = html.Div([                      # EIN Komponent, der Kinder enth√§lt
    html.Div('My First App with Data'),
    dash_table.DataTable(
        data=Dashdf.to_dict('records'),      # ganze Tabelle
        columns=[{"name": c, "id": c} for c in Dashdf.columns],
        page_size=10
    )
])

# Run the app
if __name__ == '__main__':
    #app.run(debug=True)
    #app.run_server(debug=True)
    app.run_server(host="0.0.0.0", port=8051, debug=False, use_reloader=False)
```
